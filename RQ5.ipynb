```python
import pandas as pd
from collections import Counter
```

```python
data_profiles = pd.read_csv('~/instagram_profiles.zip', delimiter='\t')
```

```python
## Sorting users by number of follwoer in descending order
data_profiles = data_profiles.sort_values(by='followers', ascending=False)
```

```python
data_profiles
```

```python
inf_n_posts = data_profiles['n_posts'].iat[0]
inf_profile_id = data_profiles['profile_id'].iat[0]
inf_followers = data_profiles['followers'].iat[0]
```

```python
inf_posts= []

for chunk in pd.read_csv('~/instagram_posts.zip', delimiter='\t', chunksize = 500000):
    inf_posts.append(chunk.loc[chunk['profile_id'] == inf_profile_id])
    ## We do this in order to avoid useless iterations, if inf_post has inf_n_posts elements,
    ## we stop iterating since we have all the posts that we need
    if(len(inf_posts) == inf_n_posts):
        break
        
inf_posts = pd.concat(inf_posts)
```

```python
## In our data sample, we only have 34 post over the 5760 the user Instagram posted
print(len(inf_posts))
print(inf_n_posts)
```

```python
## Here we take the column 'location_id' and, with the use of a Counter,
## we count the occurences of every location Instagram registered in, so we can take
## the most ones
location_cnt = Counter(inf_posts.loc[:,'location_id'])
most_location_id = location_cnt.most_common(5)
## As we can see here, we don't have a location that comes up more than once
most_location_id
```

```python
## Only keep the ids
most_location_id = [loc_id for loc_id, i in most_location_id]
most_visited_locations = []
for chunk in pd.read_csv('~/instagram_locations.zip', delimiter='\t', chunksize = 500000):
    most_visited_locations.append(chunk.loc[chunk['id'].isin(most_location_id)])
    ## We do this in order to avoid useless iterations, when we found the locations we need we stop the cycle
    if(len(most_visited_location) == 5 ):
        break
```

```python
most_visited_locations = pd.concat(most_visited_locations)
most_visited_locations
```

```python
## Counter init
post_type_counter = Counter({1:0, 2:0, 3:0})
## List init such that like_by_type[x] = number of likes of type x, like_by_type[0] will always be empty
like_by_type = [0.0, 0.0, 0.0, 0.0]
for post in inf_posts.itertuples():
    post_type_counter[post.post_type]+=1
    like_by_type[post.post_type]+=post.numbr_likes
```

```python
photo_percentage = round(post_type_counter[1] * 100/len(inf_posts),4)
video_percentage = round(post_type_counter[2] * 100/len(inf_posts),4)
multy_percentage = round(post_type_counter[3] * 100/len(inf_posts),4)
print('Post containing only photos: {}%'.format(photo_percentage))
print('Post containing only videos: {}%'.format(video_percentage))
print('Post containing both: {}%'.format(multy_percentage))

## Based on the small posts sample we have, we can say that around 91,18% of Instagram's posts are just photos,
## the rest are videos only posts and he never posts mixed posts.
```

```python
photo_likes_avg = int(round(like_by_type[1] / post_type_counter[1], 0))
videos_likes_avg = int(round(like_by_type[2] / post_type_counter[2], 0))

print('On average, post containing only photos get {} likes.'.format(photo_likes_avg))
print('On average, post containing only videos get {} likes.'.format(videos_likes_avg))

perc_photo_likers = round(100*photo_likes_avg/inf_followers, 4)
perc_video_likers = round(100*videos_likes_avg/inf_followers, 4)

print('On average, {}% of the followers likes post containing only photos.'.format(perc_photo_likers))
print('On average, {}% of the followers likes post containing only video.'.format(perc_video_likers))

## Based on our analysis of the small sample we have, we can say that just a small percentage of followers
## is actually following Instagram's account actively. In fact, likes percentage does not reach 0.5% in any
## of the cases. Also, photos' likes are slightly higher that videos' ones.
```
